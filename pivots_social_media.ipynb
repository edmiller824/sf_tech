```python
"""
create table using social media addiction table:
axis = decile of avg_daily_usage_hours
columns = bucket of addicted_score (manually created)
values = the mean of sleep_hours_per_night
"""
```

```python
import pandas as pd
import numpy as np
print('complete')
```

```python
path = "/Users/emiller/Documents/sf_tech/sf_datasets/"

sma = pd.read_csv(path + "students_social_media_addiction.csv")
sma.columns = sma.columns.str.lower()
sma.head()
```

```python
data_decile = pd.qcut(sma["avg_daily_usage_hours"], q=10, labels=False, duplicates="drop")

data_decile.name = "usage_decile"
data_decile
```

```python
sma_usage = sma.join(data_decile)
sma_usage
```

```python
# want to have lowest addicted scores in the top left of the pivot

conditions = [
    (sma_usage["addicted_score"] >= 8)
    , (sma_usage["addicted_score"] < 8) & (sma_usage["addicted_score"] >= 5) 
    , (sma_usage["addicted_score"] < 5)
]
values = ["as_bucket3", "as_bucket2", "as_bucket1"]

# conditions
sma_usage["addicted_score_bucket"] = np.select(conditions, values, default ="Other")
sma_usage
```

```python
sma_usage.pivot_table(
    index='usage_decile',
    columns='addicted_score_bucket',
    values='sleep_hours_per_night',
    aggfunc='mean',
    fill_value=0,

)
```
