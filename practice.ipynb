```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(0)

# Generate dataset
df = pd.DataFrame({
    'hours_studied': np.random.normal(5, 1, 100),
    'sleep_hours': np.random.normal(7, 0.5, 100),
    'coffee_intake': np.random.normal(2, 1, 100),
    'grades': np.random.normal(75, 10, 100)
})

# Correlation matrix
correlation = df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (without 'focus_score')")
plt.tight_layout()
plt.show()
```

```python
# sma.groupby("addicted_score").addicted_score.max()


data_decile = pd.qcut(sma["avg_daily_usage_hours"], q=10, labels=False, duplicates="drop")

data_decile.name = "usage_decile"
data_decile
# pd.qcut(sma["avg_daily_usage_hours"], q=10, labels=True, duplicates="drop")
# sma["avg_daily_usage_hours"].describe()
```

```python
sma_usage = sma.join(data_decile)
```

```python
sma
```

```python
sma_usage.groupby("usage_decile").agg({ "sleep_hours_per_night":["mean","min","max"]})
```

```python

conditions = [
    (sma_usage["addicted_score"] >= 8)
    , (sma_usage["addicted_score"] < 8) & (sma_usage["addicted_score"] >= 5) 
    , (sma_usage["addicted_score"] < 5)
]
values = ["as_bucket3", "as_bucket2", "as_bucket1"]

# conditions
sma_usage["addicted_score_bucket"] = np.select(conditions, values, default ="Other")


```
